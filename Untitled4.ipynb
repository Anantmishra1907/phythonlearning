# Seaborn


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {'Name': ['ANSH', 'SAHIL', 'JAYAN', 'ANURAG'], 'Age': [21, 23, 20, 24]}
df = pd.DataFrame(data)

sns.lineplot(x=df.index, y='Age', data=df)
plt.xlabel('Index')
plt.ylabel('Age')
plt.title('Age Line Plot')
plt.show()
```

```python
sns.barplot(x='Name', y='Age', data=df)
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Bar Plot')
plt.show()
```

```python
sns.scatterplot(x='Name', y='Age', data=df)
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Scatter Plot')
plt.show()
```

```python
sns.histplot(x='Age', data=df)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Histogram')
plt.show()
#
```

```python
sns.pointplot(x='Name', y='Age', data=df)
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Point Plot')
plt.show()
#
```

```python

sns.set_style('whitegrid')
sns.countplot(x='Name', data=df)
plt.xlabel('Name')
plt.ylabel('Count')
plt.title('Name Count Plot')
plt.show()

sns.boxplot(x='Name', y='Age', data=df)
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Box Plot')
plt.show()
#
```
