# Checking Missing Values in Pandas

```python
import pandas as pd
import numpy as np
```

```python
d = {
    'Name':['Abs ' , 'Bcd ' , np.nan],
    'Score' : [12 , np.nan , 14 ]}
df = pd.DataFrame(d)
mv = df.isnull()
print(df)
print(mv)
print(df.isna())
```

```python
nmv = df.notnull()
```

```python
print(nmv)
```
